## Data Visualization - Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))
sns.countplot(data=df_election, x='Party', order=df_election['Party'].value_counts().index)

plt.xticks(rotation=90)
plt.title("Number of Records for Each Political Party")
plt.xlabel("Party")
plt.ylabel("Count")
plt.show()
```

### Histograms

```python
plt.figure(figsize=(16,12))
for i in range(len(num_cols)):
  plt.subplot(4, 4, i+1)
  sns.histplot(df_election[num_cols[i]], kde=True, bins=30)
  plt.title(num_cols[i])
plt.tight_layout()
plt.show()
```

### Box-Plots

```python
plt.figure(figsize=(16,12))
for i in range(len(num_cols)):
  plt.subplot(4, 4, i+1)
  sns.boxplot(df_election[num_cols[i]])
  plt.title(num_cols[i])
plt.tight_layout()
plt.show()
```

### Scatterplots

```python
import matplotlib.pyplot as plt

num_cols = df_election.select_dtypes(include=['int64', 'float64']).columns

for i in range(len(num_cols)):
    for j in range(i + 1, len(num_cols)):
        plt.figure(figsize=(6, 4))
        plt.scatter(df_election[num_cols[i]], df_election[num_cols[j]])
        plt.title(f"{num_cols[i]} vs {num_cols[j]}")
        plt.xlabel(num_cols[i])
        plt.ylabel(num_cols[j])
        plt.show()
```
